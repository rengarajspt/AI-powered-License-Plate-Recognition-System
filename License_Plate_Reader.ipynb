## **Reading plates**

```python
import cv2
import numpy as np
import easyocr
import re

reader = easyocr.Reader(['en'])
# reader = easyocr.Reader(['en','ta','hi','te','kn'])

def read_plate_perfect(plate):

    # enlarge plate
    plate = cv2.resize(plate, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)

    gray = cv2.cvtColor(plate, cv2.COLOR_BGR2GRAY)

    # sharpen
    kernel = np.array([[0,-1,0],
                       [-1,5,-1],
                       [0,-1,0]])
    sharp = cv2.filter2D(gray,-1,kernel)

    # threshold
    thresh = cv2.adaptiveThreshold(
        sharp,255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        11,2
    )

    # run OCR on multiple versions
    imgs = [plate, gray, sharp, thresh]

    candidates = []

    for img in imgs:

        result = reader.readtext(
            img,
            allowlist='ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789',
            paragraph=False
        )

        if result:
            text = "".join([r[1] for r in result])
            conf = np.mean([r[2] for r in result])
            candidates.append((text, conf))

    if not candidates:
        return None

    # choose best confidence
    candidates = sorted(candidates, key=lambda x:x[1], reverse=True)
    text = candidates[0][0]

    # cleanup
    text = text.replace(" ","").upper()

    # fix common OCR mistakes
    corrections = {"O":"0","I":"1","S":"5","B":"8"}
    for k,v in corrections.items():
        text = text.replace(k,v)

    # format pattern (J389 NLT style)
    pattern = r"([A-Z]\d{3})([A-Z]{3})"
    match = re.match(pattern,text)

    if match:
        return match.group(1)+" "+match.group(2)

    return text
```
